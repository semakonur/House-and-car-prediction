### Web Scraping

```python
import pandas as pd
from selenium import webdriver

def list_to_string(s):
    return " ".join(s)

def get_data(browser, index):
    click_element = browser.find_element_by_xpath(f"//*[@id='__next']/div[3]/div[1]/div/div[5]/div[1]/div[1]/div[{index}]")
    click_element.click()

    elements = browser.find_elements_by_css_selector("._2xeaw")
    prices = browser.find_elements_by_css_selector("._34630")

    details = [element.text for element in elements]
    price = [p.text for p in prices]

    details_str = list_to_string(details)
    separated_details = details_str.split("\n")

    df = pd.DataFrame(separated_details)
    df_new = df.iloc[7:53]
    df_new.iloc[45] = df_new.iloc[45].str[0:3]

    df_new = df_new.reset_index()
    df_new.drop("index", axis=1, inplace=True)
    df_list = df_new.values.tolist()

    content = [df_list[i] for i in range(1, 46, 2)]

    price_cleaned = price[1].strip().replace("TL", "")
    content.append([price_cleaned])
    df_content = pd.DataFrame(content).T

    df_content.to_csv(r"data.csv", encoding="utf-8", index=False, mode="a")

    browser.execute_script("window.history.go(-1)")

def main():
    driver_path = "D:\PROGRAM SETUP\chromedriver.exe"
    browser = webdriver.Chrome(driver_path)
    browser.get("https://www.emlakjet.com/satilik-konut/istanbul-sariyer/3/")

    indices = list(range(1, 3)) + list(range(4, 8)) + list(range(9, 11)) + list(range(12, 14)) + list(range(15, 31))

    for index in indices:
        get_data(browser, index)

if __name__ == "__main__":
    main()
```
